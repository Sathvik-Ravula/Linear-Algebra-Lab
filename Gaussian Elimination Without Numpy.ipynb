```python
def gaussianElimination(A):
    m = len(A)
    x = all([len(row) == m + 1 for row in A[1:]])
    
    if not x:
        print("Matrix rows have unequal length")
        
    n = m + 1
    
    for k in range(m):
        #print("A[i][k] = ",A[i][k])
        pivots = [abs(A[i][k]) for i in range(k, m)]
        #print("pivots :",pivots)
        i_max = pivots.index(max(pivots)) + k
        
        #print("i_max = ",i_max)
        
        # Check for singular matrix
        if A[i_max][k] == 0:
            print("Matrix is singular!")
        
        # Swap rows
        A[k], A[i_max] = A[i_max], A[k]

        
        for i in range(k + 1, m):
            #print("\nIteration : ",i)
            f = A[i][k] / A[k][k]
            #print("\n",A)
            for j in range(k + 1, n):
                A[i][j] = A[i][j] - A[k][j] * f
            A[i][k] = 0
            
            
    sol = []
    for i in range(m - 1, -1, -1):
        sol.insert(0, A[i][m] / A[i][i])
        for k in range(i - 1, -1, -1):
            A[k][m] = A[k][m] - A[k][i] * sol[0]
        #print("\n",A)
    return sol


n = int(input("Enter No. Of Unknowns :"))
e = int(input("Enter No. Of Equations :"))

matrix = [[0 for j in range(n+1)] for i in range(e)]
#print(matrix)
for i in range(e):
    print("Equation",i+1)
    for j in range(n+1):
        if j == n:
            print("Enter the coefficient :")
            matrix[i][j] = int(input())
        else:
            print("Enter the value at ",i,j)
            matrix[i][j] = int(input())
        
#print(matrix)


system = [[1, 1, 1, 1, 1],
          [1, 1, 3, 3, 3],
          [1, 1, 2, 3, 3],
          [1, 3, 3, 3, 4]]

print("\n The Solution For The System Of Equations Is :\n",gaussianElimination(matrix))
```
